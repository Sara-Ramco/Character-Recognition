```python
import numpy as np
import pandas as pd

from keras.layers import Dense, Activation
from keras.models import Model

train_file = './train.csv'
df = pd.read_csv(train_file)

df['Pred_Int'] = df['Prediction'].apply(lambda x: ord(x) - 96)
df['Next_Pred'] = df['Prediction'].shift(-1)
df['Next_Pred'] = np.where(df['NextId'] == -1, '0', df['Next_Pred'])
df['Next_Pred_Int'] = df['Next_Pred'].apply(lambda x: ord(x) - 96)
df['Next_Pred_Int'] = np.where(df['Next_Pred_Int'] == -48, 0, df['Next_Pred_Int'])

df['Prev_Pred'] = df['Prediction'].shift(1)
df['Prev_Index'] = df['NextId'].shift(1)
df['Prev_Pred'] = np.where(df['Prev_Index'] == -1, '0', df['Prev_Pred'])
df['Prev_Pred'] = np.where(df['Prev_Index'].isnull(), '0', df['Prev_Pred'])

df['Prev_Pred_Int'] = df['Prev_Pred'].apply(lambda x: ord(x) - 96)
df['Prev_Pred_Int'] = np.where(df['Prev_Pred_Int'] == -48, 0, df['Prev_Pred_Int'])

df = df.drop(['Prev_Index'], axis =1)

#df = df.select_dtypes(['number'])
#df['Pred_Int'] = df['Pred_Int'].apply(lambda x: x/df['Pred_Int'].max())
#df['Next_Pred_Int'] = df['Next_Pred_Int'].apply(lambda x: x/df['Next_Pred_Int'].max())
#df['Prev_Pred_Int'] = df['Prev_Pred_Int'].apply(lambda x: x/df['Prev_Pred_Int'].max())
#df['Position'] = df['Position'].apply(lambda x: x/df['Position'].max())


from keras.utils import to_categorical
df['Prediction']
train_labels = to_categorical(df['Prediction'])
#cols = df.columns.tolist()
#cols.insert(0, cols.pop(cols.index('Pred_Int')))
#df = df.reindex(columns= cols)
train_labels
```

```python
X = []
for i in range(df.shape[0]):
    X_tmp = list(df.iloc[i, 3:])
    X.append(X_tmp)
data = np.array(X)
data
```

```python
target = list(df['Pred_Int'])
target
```

```python
data = data.astype('float32')
```

```python
from keras.models import Sequential

num_classes = 26
img_rows = 16
img_cols = 8
batch_size = 128
epoch = 20

model = Sequential()

model.add(Dense(32, activation='relu', input_shape=(131,)))
model.add(Activation('tanh'))

model.compile(optimizer='rmsprop',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
model.fit(data, target, epochs=5, batch_size=128)
```
